Visualize the history, including training / validation loss and accuracy

# Read in the pickle file

```python
import pickle
```

```python
hist_filename = "../history/session_2019-09-06[13_16_34].pkl"
```

```python
with open(hist_filename, "rb") as fin:
    hist_dict = pickle.load(fin)
```

```python
train_loss = hist_dict["train_loss"]
train_acc = hist_dict["train_acc"]
train_f1 = hist_dict["train_f1"]

valid_loss = hist_dict["valid_loss"]
valid_acc = hist_dict["valid_acc"]
valid_f1 = hist_dict["valid_f1"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epoch = np.arange(100)
```

# Visualize the loss function

```python
plt.plot(epoch, train_loss, "r", label="training loss")
plt.plot(epoch, valid_loss, "b", label="validation loss")
plt.legend()
plt.title("Train / Validation Loss")
plt.xlabel("epochs")
plt.ylabel("train/valid loss")
plt.show()
```

# Visualize the accuracy

```python
plt.plot(epoch, train_acc, "r", label="training accracy")
plt.plot(epoch, valid_acc, "b", label="validation accracy")
plt.legend()
plt.title("Train / Validation accuracy")
plt.xlabel("epochs")
plt.ylabel("train/valid accuracy")
plt.show()
```

```python
print("the max valid accuracy is at epoch", np.argmax(np.array(valid_acc))+1)
```

```python
max(valid_acc)
```

# Visualize the F1 score

```python
plt.plot(epoch, train_f1, "r", label="training f1_score")
plt.plot(epoch, valid_f1, "b", label="validation f1_score")
plt.legend()
plt.title("Train / Validation F1 score")
plt.xlabel("epochs")
plt.ylabel("train/valid f1 score")
plt.show()
```

```python
print("the max valid f1 score is at epoch", np.argmax(np.array(valid_f1))+1)
```

```python
max(valid_f1)
```

# Draw Statistics for Error Analysis

Here, we draw two things for training, valid and test: 
1. Confusion Matrix
2. Precision-recall Curve

```python
import numpy as np
import matplotlib.pyplot as plt
from scikitplot.metrics import plot_confusion_matrix, plot_precision_recall
```

### Train epoch 1

```python
import pickle
```

```python
train_stat_filename = "../statistics/session_2019-08-16[16_47_22]/train/train_1.pkl"
```

```python
with open(train_stat_filename, "rb") as fin:
    train_stat = pickle.load(fin)
```

```python
# for confusion matrix
train_true = train_stat["train_true"]
train_pred = train_stat["train_pred"]
```

```python
# use sklearn to get the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(train_true, train_pred).shape
```

```python
plot_confusion_matrix(train_true, train_pred)
plt.show()
```

```python
# for precision-recall curve
train_prob = train_stat["train_prob"]
#train_prob = np.expand_dims(train_prob, axis=1)
#train_prob = np.concatenate([1-train_prob, train_prob], axis=1)
#train_prob
```

```python
plot_precision_recall(train_true, train_prob, plot_micro=False)
plt.show()
```
