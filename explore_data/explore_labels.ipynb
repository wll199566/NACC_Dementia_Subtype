```python
# to see how many patients have both NPADNC and NACCLEWY
npadnc_nacclewy_df = nacc_csv.loc[(nacc_csv["NPADNC"].isin([0, 1, 2, 3]) & (nacc_csv["NACCLEWY"].isin([0, 1, 2, 3])))]
```

```python
len(pd.unique(npadnc_nacclewy_df["NACCID"]))
```

```python
npadnc_nacclewy_df[["NACCID", "DATE", "CDRGLOB", "NPADNC", "NACCLEWY"]]
```

```python
# get those patients whose initial visit is normal or on their early stage
qualified_init_visit_list = []
for naccid in pd.unique(npadnc_nacclewy_df["NACCID"]):
    patient_record = npadnc_nacclewy_df[npadnc_nacclewy_df["NACCID"]==naccid]
    if patient_record.head(1)["CDRGLOB"].isin([0, 0.5, 1]).values[0]:
        qualified_init_visit_list.append(naccid)
```

```python
len(qualified_init_visit_list)
```

```python
# to check those patients
qualified_init_visit_df = npadnc_nacclewy_df.loc[npadnc_nacclewy_df["NACCID"].isin(qualified_init_visit_list)]
```

```python
# to check how many the patients are whose initial visits of CDRGLOB == 0.0
init_visit_zero_list = []
for naccid in pd.unique(qualified_init_visit_df["NACCID"]):
    patient_record = qualified_init_visit_df[qualified_init_visit_df["NACCID"]==naccid]
    if patient_record.head(1)["CDRGLOB"].values[0] == 0:
        init_visit_zero_list.append(naccid)
```

```python
len(init_visit_zero_list)
```

```python
init_visit_zero_df_partial = qualified_init_visit_df[qualified_init_visit_df["NACCID"].isin(init_visit_zero_list)][["NACCID", "DATE", "CDRGLOB", "NPADNC", "NACCLEWY"]]
```

```python
init_visit_zero_list[0]
```

```python
init_visit_zero_df_partial[init_visit_zero_df_partial["NACCID"]=='NACC003627']["CDRGLOB"].values
```

```python
for naccid in init_visit_zero_list:
    patient_records = init_visit_zero_df_partial.loc[init_visit_zero_df_partial["NACCID"]==naccid]
    print(naccid)
    print(patient_records["CDRGLOB"].values)
    print('\n')
```

We can see that all those patients with initial CDRGLOB score==0, they will not convert to 2 or 3 suddenly without going through 0.5 or 1.

```python
# to get how many people have PURE AD, PURE LBD and MIXED AD+LBD
pure_ad_df = qualified_init_visit_df[(qualified_init_visit_df["NPADNC"].isin([2,3])) & (qualified_init_visit_df["NACCLEWY"]==0)]
pure_lbd_df = qualified_init_visit_df[(qualified_init_visit_df["NPADNC"].isin([0, 1])) & (qualified_init_visit_df["NACCLEWY"]==3)]
mixed_ad_lbd_df = qualified_init_visit_df[(qualified_init_visit_df["NPADNC"].isin([2,3])) & (qualified_init_visit_df["NACCLEWY"].isin([1,2,3]))]
```

```python
print("the number of PURE AD patients:", len(pd.unique(pure_ad_df["NACCID"])))
print("the number of PURE LBD patients:", len(pd.unique(pure_lbd_df["NACCID"])))
print("the number of MIXED AD + LBD patients:", len(pd.unique(mixed_ad_lbd_df["NACCID"])))
```

```python
### get the NC+Mild window from the qualified_init_visit_df ###
nc_mild_index = []

qualified_patients_cdr_df = qualified_init_visit_df[["NACCID", "DATE", "CDRGLOB"]]

# find the qualified row index
for naccid in pd.unique(qualified_patients_cdr_df["NACCID"]):
    # get the record for each patient
    patient_records = qualified_patients_cdr_df.loc[qualified_patients_cdr_df["NACCID"]==naccid]
    # iterate each index and find the qualified index
    for index, row in patient_records.iterrows():
        # once we encounter the later stage, we don't continue for this patient
        if (row[2] == 2) or (row[2] == 3):
            break
        nc_mild_index.append(index)
```

```python
nc_mild_index
```

```python
qualified_patients_nc_mild_window_df = qualified_patients_cdr_df.loc[nc_mild_index, :]
```

```python
# check the tansition matrix before we get the NC + MILD window
from explore_data import get_cdr_transitions

print(get_cdr_transitions(qualified_patients_cdr_df))
```

```python
# check the tansition matrix after we get the NC + MILD window
from explore_data import get_cdr_transitions

print(get_cdr_transitions(qualified_patients_nc_mild_window_df))
```

We can see that they cannot convert into CDRGLOB = \[2, 3\]. Thus, we succeed get NC+MILD window.

```python
# get the feature dataframe according to the index we got above
qualified_patients_features_df = qualified_init_visit_df.loc[nc_mild_index, :]
qualified_patients_features_df = qualified_patients_features_df.sort_values(by=["NACCID", "DATE"]).reset_index(drop=True)
```
