# Read in the pickle file

```python
import pickle
```

```python
#ls ../history
!ls /gpfs/data/razavianlab/Hui_NACC_hist_out_models/history/lr_origin_history/
```

```python
history_path = "/gpfs/data/razavianlab/Hui_NACC_hist_out_models/history/lr_origin_history/"
```

```python
hist_filename = history_path + "session_2019-10-08[12_39_04].pkl" # no_reg
#hist_filename = history_path + "session_2019-10-08[12_53_45].pkl" # a=0.001
#hist_filename = history_path + "session_2019-10-08[17_00_33].pkl" # a=0.01
#hist_filename = history_path + "session_2019-10-08[12_55_04].pkl" # a=0.1
```

```python
with open(hist_filename, "rb") as fin:
    hist_dict = pickle.load(fin)
```

```python
train_loss = hist_dict["train_loss"]
train_f1_macro = hist_dict["train_f1_macro"]
train_f1_micro = hist_dict["train_f1_micro"]

valid_loss = hist_dict["valid_loss"]
valid_f1_macro = hist_dict["valid_f1_macro"]
valid_f1_micro = hist_dict["valid_f1_micro"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epoch = np.arange(300)
```

# Visualize the loss function

```python
plt.plot(epoch, train_loss, "r", label="training loss")
plt.plot(epoch, valid_loss, "b", label="validation loss")
plt.legend()
plt.title("Train / Validation Loss")
plt.xlabel("epochs")
plt.ylabel("train/valid loss")
plt.show()
```

# Visualize the macro_f1

```python
plt.plot(epoch, train_f1_macro, "r", label="training macro_f1")
plt.plot(epoch, valid_f1_macro, "b", label="validation macro_f1")
plt.legend()
plt.title("Train / Validation macro f1")
plt.xlabel("epochs")
plt.ylabel("train/valid macro f1")
plt.show()
```

```python
print("the max valid accuracy is at epoch", np.argmax(np.array(valid_f1_macro))+1)
```

```python
max(valid_f1_macro)
```

# Visualize the micro_f1

```python
plt.plot(epoch, train_f1_micro, "r", label="training micro_f1")
plt.plot(epoch, valid_f1_micro, "b", label="validation micro_f1")
plt.legend()
plt.title("Train / Validation micro F1")
plt.xlabel("epochs")
plt.ylabel("train/valid micro F1")
plt.show()
```

```python
print("the max valid f1 score is at epoch", np.argmax(np.array(valid_f1_micro))+1)
```

```python
max(valid_f1_micro)
```
