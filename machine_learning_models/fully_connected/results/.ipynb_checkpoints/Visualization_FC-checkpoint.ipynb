```python
with open(hist_filename, "rb") as fin:
    hist_dict = pickle.load(fin)
```

```python
hist_dict.keys()
```

```python
train_loss = hist_dict["train_loss"]
train_f1_macro = hist_dict["train_f1_macro"]
train_f1_micro = hist_dict["train_f1_micro"]

valid_loss = hist_dict["valid_loss"]
valid_f1_macro = hist_dict["valid_f1_macro"]
valid_f1_micro = hist_dict["valid_f1_micro"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epoch = np.arange(300)
```

# Visualize the loss function

```python
plt.plot(epoch, train_loss, "r", label="training loss")
plt.plot(epoch, valid_loss, "b", label="validation loss")
plt.legend()
plt.title("Train / Validation Loss")
plt.xlabel("epochs")
plt.ylabel("train/valid loss")
plt.show()
```

# Visualize the f1 macro score

```python
plt.plot(epoch, train_f1_macro, "r", label="training f1 macro")
plt.plot(epoch, valid_f1_macro, "b", label="validation f1 macro")
plt.legend()
plt.title("Train / Validation f1 score")
plt.xlabel("epochs")
plt.ylabel("train/valid f1 score")
plt.show()
```

```python
print("the max valid f1 macro is at epoch", np.argmax(np.array(valid_f1_macro))+1)
```

```python
max(valid_f1_macro)
```

# Visualize micro F1 score

```python
plt.plot(epoch, train_f1_micro, "r", label="training f1 micro score")
plt.plot(epoch, valid_f1_micro, "b", label="validation f1 micro score")
plt.legend()
plt.title("Train / Validation micro F1 score")
plt.xlabel("epochs")
plt.ylabel("train/valid f1 score")
plt.show()
```

```python
print("the max valid f1 score is at epoch", np.argmax(np.array(valid_f1_micro))+1)
```

```python
max(valid_f1_micro)
```
